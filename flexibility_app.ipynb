# Exploring the trade-offs between flexibility and stability

Imagine we are trying to estimate some 'true' relationship between our predictor $x$ and our outcome of interest $y$. This simple app demonstrates that the most complicated (most flexible) model doesn't always do the best in terms of minimizing our error in estimating this true function.

Start the app by running the cell below ( by highlighting and typing shift+enter) to see the app. Then read the text below the plot carefully, and complete the tasks.




```python
from generate_app import FlexibilityApp
app = FlexibilityApp()
app.display()
```

The real relationship between the $x$ and $y$ is displayed by the red line in the plot - we can call this the 'true function' $f(x)$.

In practice, we observe only a limited number of noisy data points that give us some information about this true function. These data are represented by the blue dots. We can control how many datapoints we observe, and how noisy our observations are, using the sliders on the app. By noise in this context, we mean how far points are away from the red line, on average.

Next, the app fits a model to the observed data, and displays the relationship between x and y estimated by the model - this is the blue line. For the model to be useful, we want the gap between the blue and the red line to be as small as possible, so that the model captures the true underlying relationship and therefore generalises well to new data. To put it another way, a model for which the estimated function is close to the true function will be able to make good predictions on new data.

We can control the flexibility of the model we are fitting using the "Model Complexity" slider. The more complicated the model, the more closely it can approximate the data, but the more likely it is to over-fit to the specific set of observations it is estimated on (including the noise in those observations). In this case, model complexity is a *hyperparameter* that we need to tune to find the best result. Here we are manually setting the complexity hyperparameter and we know the true function used to generate the data. In practice, we would choose the hyperparameter value that results in the smallest loss for the validation dataset.
